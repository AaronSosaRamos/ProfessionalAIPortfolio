#Introducing the Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """Define a function f(x) for demonstration."""
    return x**2

def derivative(f, x, h=1e-5):
    """
    Compute the derivative of a function f at point x using the limit definition.

    Parameters:
        f (function): The function for which the derivative is to be computed.
        x (float): The point at which to compute the derivative.
        h (float): A small value used to compute the numerical derivative (default 1e-5).

    Returns:
        float: The value of the derivative of f at x.
    """
    # Use the limit definition of the derivative: f'(x) = lim(h->0) [f(x+h) - f(x)] / h
    derivative_value = (f(x + h) - f(x)) / h
    return derivative_value

# Define the point at which to compute the derivative
x_value = 2.0

# Compute the derivative of f at x_value
derivative_value = derivative(f, x_value)

# Print the computed derivative
print(f"The derivative of f at x = {x_value} is approximately: {derivative_value}")

# Plotting the function and its tangent line at x = x_value
x = np.linspace(-2, 3, 400)  # Generate x values for plotting
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2')
plt.title('Function f(x) = x^2 and its Tangent at x = 2')
plt.xlabel('x')
plt.ylabel('f(x)')

# Plot tangent line (derivative) at x = x_value
tangent_x = np.array([x_value - 1, x_value + 1])  # X values for tangent line
tangent_y = f(x_value) + derivative_value * (tangent_x - x_value)  # Y values for tangent line
plt.plot(tangent_x, tangent_y, label=f'Tangent (Derivative = {derivative_value:.2f})', linestyle='--')
plt.scatter(x_value, f(x_value), color='red')  # Mark the point (x_value, f(x_value))

plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(x):
    """Define a function f(x) for demonstration."""
    return x**2 + sp.sin(x)  # Example of a more complex function

def compute_derivative(f, x):
    """
    Compute the derivative of a function f symbolically using sympy.

    Parameters:
        f (sympy.Expr): The sympy expression representing the function.
        x (sympy.Symbol): The symbol with respect to which the derivative is computed.

    Returns:
        sympy.Expr: The symbolic expression of the derivative of f with respect to x.
    """
    return sp.diff(f, x)

# Define the symbolic variable
x = sp.symbols('x')

# Define the function f(x) symbolically
f_expr = x**2 + sp.sin(x)

# Compute the derivative of f with respect to x
f_derivative = compute_derivative(f_expr, x)

# Print the symbolic expression of the derivative
print("Symbolic derivative of f(x):")
print(f_derivative)

# Now, let's evaluate the derivative at a specific point (e.g., x = 2.0)
x_value = 2.0
derivative_value = f_derivative.subs(x, x_value)

# Print the evaluated derivative value
print(f"The derivative of f at x = {x_value} is approximately: {derivative_value}")

# Plotting the function and its tangent line at x = x_value
x_vals = np.linspace(-2, 3, 400)
y_vals = np.array([f_expr.subs(x, xv) for xv in x_vals], dtype=float)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2 + sin(x)')
plt.title('Function f(x) = x^2 + sin(x) and its Tangent at x = 2')
plt.xlabel('x')
plt.ylabel('f(x)')

# Plot tangent line (derivative) at x = x_value
tangent_x = np.array([x_value - 1, x_value + 1])
tangent_y = f_expr.subs(x, x_value) + derivative_value * (tangent_x - x_value)
plt.plot(tangent_x, tangent_y, label=f'Tangent (Derivative ≈ {derivative_value:.2f})', linestyle='--')
plt.scatter(x_value, f_expr.subs(x, x_value), color='red')  # Mark the point (x_value, f(x_value))

plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

#Rules for differentiation of all known functions


```python
import sympy as sp

# Define symbolic variables
x, a = sp.symbols('x a')

# Basic Differentiation Rules
def differentiate(f, x):
    """Compute the derivative of function f symbolically with respect to x."""
    return sp.diff(f, x)

# Differentiation Rules for Basic Functions
def differentiate_all_functions():
    # Constants
    constant_rule = differentiate(5, x)  # d/dx (c) = 0
    print("Constant Rule: d/dx (c) =", constant_rule)

    # Power Rule: d/dx (x^n) = n*x^(n-1)
    power_rule = differentiate(x**2, x)
    print("Power Rule: d/dx (x^a) =", power_rule)

    # Exponential Function: d/dx (e^x) = e^x
    exp_rule = differentiate(sp.exp(x), x)
    print("Exponential Rule: d/dx (e^x) =", exp_rule)

    # Logarithmic Function: d/dx (ln(x)) = 1/x
    log_rule = differentiate(sp.ln(x), x)
    print("Logarithmic Rule: d/dx (ln(x)) =", log_rule)

    # Trigonometric Functions
    # d/dx (sin(x)) = cos(x)
    sin_rule = differentiate(sp.sin(x), x)
    print("Sin Rule: d/dx (sin(x)) =", sin_rule)

    # d/dx (cos(x)) = -sin(x)
    cos_rule = differentiate(sp.cos(x), x)
    print("Cos Rule: d/dx (cos(x)) =", cos_rule)

    # d/dx (tan(x)) = sec^2(x)
    tan_rule = sp.diff(sp.tan(x), x)
    print("Tan Rule: d/dx (tan(x)) =", tan_rule)

    # Combining Rules: Example: d/dx (3x^2 + 2sin(x) - ln(x))
    combined_function = 3*x**2 + 2*sp.sin(x) - sp.ln(x)
    combined_rule = differentiate(combined_function, x)
    print("Differentiation of 3x^2 + 2sin(x) - ln(x):", combined_rule)

# Call the function to demonstrate all differentiation rules
differentiate_all_functions()
```

```python
import sympy as sp

# Define symbolic variable
x = sp.symbols('x')

# Function to compute and print differentiation rules
def differentiate_all_functions():
    # Constants
    constant_rule = sp.diff(5, x)  # d/dx (c) = 0
    print("Constant Rule: d/dx (c) =", constant_rule)

    # Power Rule: d/dx (x^n) = n*x^(n-1)
    power_rule = sp.diff(x**3, x)
    print("Power Rule: d/dx (x^3) =", power_rule)

    # Exponential Function: d/dx (e^x) = e^x
    exp_rule = sp.diff(sp.exp(x), x)
    print("Exponential Rule: d/dx (e^x) =", exp_rule)

    # Logarithmic Function: d/dx (ln(x)) = 1/x
    log_rule = sp.diff(sp.ln(x), x)
    print("Logarithmic Rule: d/dx (ln(x)) =", log_rule)

    # Trigonometric Functions
    # d/dx (sin(x)) = cos(x)
    sin_rule = sp.diff(sp.sin(x), x)
    print("Sin Rule: d/dx (sin(x)) =", sin_rule)

    # d/dx (cos(x)) = -sin(x)
    cos_rule = sp.diff(sp.cos(x), x)
    print("Cos Rule: d/dx (cos(x)) =", cos_rule)

    # d/dx (tan(x)) = sec^2(x)
    tan_rule = sp.diff(sp.tan(x), x)
    print("Tan Rule: d/dx (tan(x)) =", tan_rule)

    # Hyperbolic Functions
    # d/dx (sinh(x)) = cosh(x)
    sinh_rule = sp.diff(sp.sinh(x), x)
    print("Hyperbolic Sin Rule: d/dx (sinh(x)) =", sinh_rule)

    # d/dx (cosh(x)) = sinh(x)
    cosh_rule = sp.diff(sp.cosh(x), x)
    print("Hyperbolic Cos Rule: d/dx (cosh(x)) =", cosh_rule)

    # d/dx (tanh(x)) = sech^2(x)
    tanh_rule = sp.diff(sp.tanh(x), x)
    print("Hyperbolic Tan Rule: d/dx (tanh(x)) =", tanh_rule)

    # Inverse Trigonometric Functions
    # d/dx (arcsin(x)) = 1 / sqrt(1 - x^2)
    arcsin_rule = sp.diff(sp.asin(x), x)
    print("Inverse Sin Rule: d/dx (arcsin(x)) =", arcsin_rule)

    # d/dx (arccos(x)) = -1 / sqrt(1 - x^2)
    arccos_rule = sp.diff(sp.acos(x), x)
    print("Inverse Cos Rule: d/dx (arccos(x)) =", arccos_rule)

    # d/dx (arctan(x)) = 1 / (1 + x^2)
    arctan_rule = sp.diff(sp.atan(x), x)
    print("Inverse Tan Rule: d/dx (arctan(x)) =", arctan_rule)

    # Logarithmic Function: d/dx (log(x)) = 1/x
    log_rule = sp.diff(sp.log(x), x)
    print("Logarithmic Rule: d/dx (log(x)) =", log_rule)

    # Stepwise Function (Heaviside Step Function)
    # d/dx (u(x)) = 0 for x != 0, undefined at x = 0 (Dirac delta function)
    stepwise_rule = sp.diff(sp.Heaviside(x), x)
    print("Stepwise Function Rule: d/dx (Heaviside(x)) =", stepwise_rule)

# Call the function to demonstrate all differentiation rules
differentiate_all_functions()
```

```python
import sympy as sp

# Define symbolic variable
x = sp.symbols('x')

# Define different types of functions
# Power function: f(x) = x^3
power_function = x**3

# Sum of two functions: f(x) = sin(x) + x^2
sum_function = sp.sin(x) + x**2

# Difference of two functions: f(x) = exp(x) - cos(x)
difference_function = sp.exp(x) - sp.cos(x)

# Composite function: f(x) = sinh(x^2)
composite_function = sp.sinh(x**2)

# Product of two functions: f(x) = x * exp(x)
product_function = x * sp.exp(x)

# Constant multiple of a function: f(x) = 2 * ln(x)
constant_multiple_function = 2 * sp.ln(x)

# Quotient of two functions: f(x) = tan(x) / x
quotient_function = sp.tan(x) / x

# Function with stepwise behavior: Heaviside step function
stepwise_function = sp.Heaviside(x)

# Function to compute and print differentiation rules for different functions
def differentiation_examples():
    # Power Rule: d/dx (x^3) = 3*x^2
    power_derivative = sp.diff(power_function, x)
    print("Power Rule (d/dx (x^3)) =", power_derivative)

    # Sum Rule: d/dx (sin(x) + x^2) = cos(x) + 2*x
    sum_derivative = sp.diff(sum_function, x)
    print("Sum Rule (d/dx (sin(x) + x^2)) =", sum_derivative)

    # Difference Rule: d/dx (e^x - cos(x)) = e^x + sin(x)
    difference_derivative = sp.diff(difference_function, x)
    print("Difference Rule (d/dx (e^x - cos(x))) =", difference_derivative)

    # Chain Rule: d/dx (sinh(x^2)) = 2x*cosh(x^2)
    composite_derivative = sp.diff(composite_function, x)
    print("Chain Rule (d/dx (sinh(x^2))) =", composite_derivative)

    # Product Rule: d/dx (x * e^x) = e^x + x*e^x
    product_derivative = sp.diff(product_function, x)
    print("Product Rule (d/dx (x * e^x)) =", product_derivative)

    # Constant Multiple Rule: d/dx (2*ln(x)) = 2/x
    constant_multiple_derivative = sp.diff(constant_multiple_function, x)
    print("Constant Multiple Rule (d/dx (2*ln(x))) =", constant_multiple_derivative)

    # Quotient Rule: d/dx (tan(x) / x) = (sec^2(x)*x - tan(x)) / x^2
    quotient_derivative = sp.diff(quotient_function, x)
    print("Quotient Rule (d/dx (tan(x) / x)) =", quotient_derivative)

    # Derivative of Heaviside Step Function
    # Note: Derivative of Heaviside step function is Dirac delta function (undefined at x = 0)
    stepwise_derivative = sp.diff(stepwise_function, x)
    print("Derivative of Heaviside Step Function (d/dx (Heaviside(x))) =", stepwise_derivative)

# Call the function to demonstrate differentiation rules for different functions
differentiation_examples()
```

```python
import sympy as sp

# Define symbolic variable and function
x = sp.symbols('x')
f = sp.sin(x)

# Compute Taylor series expansion around x = 0 up to 4th order
taylor_series = f.series(x, 0, 5).removeO()  # Expand up to x^4 term
print("Taylor Series Expansion of sin(x) around x = 0 up to 4th order:")

# Explanation of each step
order = 4  # Highest order term in the expansion

# Iterate over each order from 0 to 'order'
for n in range(order + 1):
    # Compute nth derivative of sin(x) at x = 0
    nth_derivative = sp.diff(f, x, n).subs(x, 0)

    # Compute nth term in the Taylor series expansion
    nth_term = (nth_derivative / sp.factorial(n)) * (x**n)

    # Print the nth term of the Taylor series
    print(f"Order {n}: {nth_term}")

# Print the full Taylor series expansion up to 4th order
print("Full Taylor Series Expansion (up to x^4):", taylor_series)
```

```python
import numpy as np

def f(x):
    """Define the function f(x) for numerical differentiation."""
    return np.sin(x)

def centered_difference(f, x, h=1e-5):
    """
    Compute the derivative of function f at point x using centered difference scheme.

    Parameters:
        f (function): The function for which the derivative is to be computed.
        x (float): The point at which to compute the derivative.
        h (float): The step size for numerical differentiation (default 1e-5).

    Returns:
        float: The approximated derivative of f at x.
    """
    return (f(x + h) - f(x - h)) / (2 * h)

# Compute derivative of sin(x) at x = 1 using centered difference
x_value = 1.0
approx_derivative = centered_difference(f, x_value)
print("Approximated derivative of sin(x) at x = 1:", approx_derivative)
```
