```python
def limit(f, a):
    h = 1e-9  # Small value for numerical approximation
    return (f(a + h) - f(a)) / h

# Example usage:
# Define a function f(x) that we want to find the limit of
def f(x):
    return x**2

# Compute the limit of f(x) as x approaches 2
limit_value = limit(f, 2)
print("Limit of f(x) as x approaches 2:", limit_value)
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable and function symbols
x = sp.symbols('x')
f = sp.Function('f')(x)
g = sp.Function('g')(x)

# Define example functions for f(x) and g(x)
f_expr = x**2 - 3*x + 2
g_expr = 2*x + 1

# Define specific value for a (the limit point)
a = 3

# Evaluate limits at the point 'a'
limit_f_at_a = sp.limit(f_expr, x, a)
limit_g_at_a = sp.limit(g_expr, x, a)

# Calculate limit laws using specific value of 'a'
sum_rule = sp.limit(f_expr + g_expr, x, a)
diff_rule = sp.limit(f_expr - g_expr, x, a)
const_mult_rule = sp.limit(3 * f_expr, x, a)
product_rule = sp.limit(f_expr * g_expr, x, a)
quotient_rule = sp.limit(f_expr / g_expr, x, a)

# Print the results of the limits and limit laws
print(f"Limit of f(x) at x = {a}: {limit_f_at_a}")
print(f"Limit of g(x) at x = {a}: {limit_g_at_a}")
print("Sum Rule:", sum_rule)
print("Difference Rule:", diff_rule)
print("Constant Multiple Rule:", const_mult_rule)
print("Product Rule:", product_rule)
print("Quotient Rule:", quotient_rule)

# Plotting example functions and their limits around 'a'
x_vals = np.linspace(a - 3, a + 3, 400)  # Range around 'a'
f_vals = sp.lambdify(x, f_expr)(x_vals)
g_vals = sp.lambdify(x, g_expr)(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = x^2 - 3x + 2')
plt.plot(x_vals, g_vals, label='g(x) = 2x + 1')
plt.plot(a, limit_f_at_a, 'ro', label=f'Limit of f(x) at x={a}')
plt.plot(a, limit_g_at_a, 'bo', label=f'Limit of g(x) at x={a}')
plt.title('Example Functions and Limits')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.axvline(a, color='gray', linestyle='--', linewidth=1)  # Vertical line at 'a'

plt.show()
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable and function symbols
x = sp.symbols('x')
f = sp.Function('f')(x)
g = sp.Function('g')(x)

# Define example functions for f(x) and g(x)
f_expr = sp.sin(x) / x  # Example: sin(x) / x
g_expr = sp.exp(x)  # Example: e^x

# Define specific value for a (the limit point)
a = 3  # We'll evaluate limits at x = 0 for these examples

# Evaluate limits at the point 'a'
limit_f_at_a = sp.limit(f_expr, x, a)
limit_g_at_a = sp.limit(g_expr, x, a)

# Calculate various limit laws using specific value of 'a'
sum_rule = sp.limit(f_expr + g_expr, x, a)
diff_rule = sp.limit(f_expr - g_expr, x, a)
const_mult_rule = sp.limit(3 * f_expr, x, a)
product_rule = sp.limit(f_expr * g_expr, x, a)
quotient_rule = sp.limit(f_expr / g_expr, x, a)

# Composition Rule: lim(x->a) [f(g(x))] = f(lim(x->a) g(x))
composition_rule = sp.limit(f.subs(x, g_expr), x, a)

# Exponential and Logarithmic Limits
exp_limit = sp.limit(sp.exp(f_expr), x, a)
log_limit = sp.limit(sp.log(g_expr), x, a)

# Print the results of the limits and limit laws
print(f"Limit of f(x) at x = {a}: {limit_f_at_a}")
print(f"Limit of g(x) at x = {a}: {limit_g_at_a}")
print("Sum Rule:", sum_rule)
print("Difference Rule:", diff_rule)
print("Constant Multiple Rule:", const_mult_rule)
print("Product Rule:", product_rule)
print("Quotient Rule:", quotient_rule)
print("Composition Rule:", composition_rule)
print("Exponential Limit:", exp_limit)
print("Logarithmic Limit:", log_limit)

# Plotting example functions and their limits around 'a'
x_vals = np.linspace(-5, 5, 400)  # Range around 'a'
f_vals = sp.lambdify(x, f_expr)(x_vals)
g_vals = sp.lambdify(x, g_expr)(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = sin(x) / x')
plt.plot(x_vals, g_vals, label='g(x) = e^x')
plt.plot(a, limit_f_at_a, 'ro', label=f'Limit of f(x) at x={a}')
plt.plot(a, limit_g_at_a, 'bo', label=f'Limit of g(x) at x={a}')
plt.title('Example Functions and Limits')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.axvline(a, color='gray', linestyle='--', linewidth=1)  # Vertical line at 'a'

plt.show()
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable and function symbol
x = sp.symbols('x')
f = sp.Function('f')(x)

# Define a piecewise function for f(x)
f_expr = sp.Piecewise((x**2, x < 2), (4, x >= 2))  # f(x) = x^2 for x < 2, f(x) = 4 for x >= 2

# Define the point where we want to check continuity
c = 2

# Evaluate the function and limit at the point of interest
value_at_c = f_expr.subs(x, c)
limit_at_c = sp.limit(f_expr, x, c)

# Check if the function is defined at the point c
is_defined_at_c = sp.limit(f_expr, x, c).is_finite

# Plotting the function and indicating the point of interest
x_vals_left = np.linspace(1, 2, 400)  # Range for x < 2
x_vals_right = np.linspace(2, 3, 400)  # Range for x >= 2
f_vals_left = [sp.lambdify(x, f_expr.args[0][0])(xi) for xi in x_vals_left]  # Evaluate f(x) = x^2 for x < 2
f_vals_right = [sp.lambdify(x, f_expr.args[1][0])(xi) for xi in x_vals_right]  # Evaluate f(x) = 4 for x >= 2

plt.figure(figsize=(10, 6))
plt.plot(x_vals_left, f_vals_left, label='f(x) = x^2 for x < 2')
plt.plot(x_vals_right, f_vals_right, label='f(x) = 4 for x >= 2')
plt.plot(c, value_at_c, 'ro', label=f'Value at x={c}: {value_at_c}')
plt.axvline(c, color='gray', linestyle='--', linewidth=1, label=f'Point x={c}')
plt.title('Piecewise Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Print the results of continuity check
print("Function is defined at x = {}? {}".format(c, is_defined_at_c))
print("Value of the function at x = {}: {}".format(c, value_at_c))
print("Limit of the function at x = {}: {}".format(c, limit_at_c))
print("Is the function continuous at x = {}? {}".format(c, value_at_c == limit_at_c))
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variable and function symbol
x = sp.symbols('x')
f = sp.Function('f')(x)

# Define a piecewise function with a jump discontinuity at x = 1
f_expr = sp.Piecewise((x**2, x < 1), (3, x > 1))  # f(x) = x^2 for x < 1, f(x) = 3 for x > 1

# Define the point where we want to evaluate the limit
a = 1

# Evaluate the function and limit at the discontinuity point
value_at_a = f_expr.subs(x, a)
limit_at_a_left = sp.limit(f_expr, x, a, dir='-')  # Limit from the left (approaching from x < 1)
limit_at_a_right = sp.limit(f_expr, x, a, dir='+')  # Limit from the right (approaching from x > 1)

# Plotting the function and indicating the point of interest
x_vals_left = np.linspace(0, 1, 400)  # Range for x < 1
x_vals_right = np.linspace(1, 2, 400)  # Range for x > 1
f_vals_left = [sp.lambdify(x, f_expr.args[0][0])(xi) for xi in x_vals_left]  # Evaluate f(x) = x^2 for x < 1
f_vals_right = [sp.lambdify(x, f_expr.args[1][0])(xi) for xi in x_vals_right]  # Evaluate f(x) = 3 for x > 1

plt.figure(figsize=(10, 6))
plt.plot(x_vals_left, f_vals_left, label='f(x) = x^2 for x < 1')
plt.plot(x_vals_right, f_vals_right, label='f(x) = 3 for x > 1')
plt.plot(a, value_at_a, 'ro', label=f'Value at x={a}: {value_at_a}')
plt.axvline(a, color='gray', linestyle='--', linewidth=1, label=f'Discontinuity at x={a}')
plt.title('Piecewise Function with Jump Discontinuity')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Print the results of the limit evaluations
print("Value of the function at x = {}: {}".format(a, value_at_a))
print("Limit of the function at x = {} from the left: {}".format(a, limit_at_a_left))
print("Limit of the function at x = {} from the right: {}".format(a, limit_at_a_right))
print("Is the limit at x = {} existent? {}".format(a, limit_at_a_left == limit_at_a_right))
```

#Intermediate Value Theorem

```python
def find_ivt_point(f, a, b, L, tolerance=1e-6, max_iterations=1000):
    """
    Find a point `c` in the interval [a, b] such that f(c) is approximately L
    using the Intermediate Value Theorem.

    Args:
        f: The function for which to find the point.
        a: The left endpoint of the interval.
        b: The right endpoint of the interval.
        L: The target value that f(c) should be close to.
        tolerance: The acceptable error tolerance for |f(c) - L|.
        max_iterations: Maximum number of iterations for the numerical method.

    Returns:
        c: The value within [a, b] such that f(c) is close to L,
           or None if no such c is found within the maximum iterations.
    """
    if f(a) > f(b):
        a, b = b, a  # Ensure f(a) <= f(b) to cover both increasing and decreasing cases

    for _ in range(max_iterations):
        c = (a + b) / 2  # Take midpoint of the current interval [a, b]
        fc = f(c)

        if abs(fc - L) < tolerance:
            return c  # Found a point c where f(c) is close to L

        # Update the interval [a, b] based on the sign of f(c) - L
        if fc < L:
            a = c  # Move left endpoint to c
        else:
            b = c  # Move right endpoint to c

    return None  # If no suitable c is found within max_iterations

# Example usage:
import math

# Define the function for which we want to apply the Intermediate Value Theorem
def f(x):
    return x**3 - x**2 + 2*x - 1

# Define the interval [a, b] and the target value L
a = -2
b = 2
L = 0

# Find a point c where f(c) is close to L using the Intermediate Value Theorem
c = find_ivt_point(f, a, b, L)
if c is not None:
    print(f"Found c = {c} such that f(c) is approximately {L}")
    print(f"f({c}) = {f(c)}")
else:
    print("No suitable point c found within the specified iterations.")
```
