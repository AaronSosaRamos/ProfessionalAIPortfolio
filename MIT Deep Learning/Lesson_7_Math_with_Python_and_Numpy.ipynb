#sigmoid function, np.exp()

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function definition
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)  # Generate 100 points between -10 and 10

# Compute sigmoid values
y = sigmoid(x)

# Plot sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sigmoid(x)', color='b')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.axhline(0.5, color='r', linestyle='--', linewidth=0.8)  # Add y=0.5 line for reference
plt.legend()
plt.show()
```

image.png

```python
# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Compute sigmoid derivative values
dy = sigmoid_derivative(x)

# Plot sigmoid function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sigmoid(x)', color='b')
plt.plot(x, dy, label="sigmoid'(x)", color='g', linestyle='--')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np

# Sigmoid function definition
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Generate a vector of data
data = np.array([1.0, 2.0, 0.5, -1.0, -2.0])

# Apply sigmoid gradient (sigmoid derivative) to the vector element-wise
sigmoid_gradient_data = sigmoid_derivative(data)

print("Original Data:", data)
print("Sigmoid Gradient (Sigmoid Derivative) applied to data:", sigmoid_gradient_data)
```

#Min-Max Scaling

```python
import numpy as np

# Generate a synthetic dataset (10 rows x 5 columns)
np.random.seed(0)
dataset1 = np.random.rand(10, 5)  # Random values between 0 and 1

# Min-Max Scaling (scaling each row to [0, 1] range)
min_vals = np.min(dataset1, axis=1, keepdims=True)
max_vals = np.max(dataset1, axis=1, keepdims=True)
normalized_dataset1 = (dataset1 - min_vals) / (max_vals - min_vals)

print("Dataset 1 (before normalization):")
print(dataset1)
print("\nDataset 1 (after Min-Max Scaling):")
print(normalized_dataset1)
```

#Standardization (Z-score Scaling)

```python
# Standardization (Z-score Scaling)
mean_vals = np.mean(dataset1, axis=1, keepdims=True)
std_vals = np.std(dataset1, axis=1, keepdims=True)
standardized_dataset1 = (dataset1 - mean_vals) / std_vals

print("\nDataset 1 (after Standardization):")
print(standardized_dataset1)
```

#L2 Normalization

```python
# L2 Normalization (scaling each row to have unit norm)
norms = np.linalg.norm(dataset1, axis=1, keepdims=True)
normalized_l2_dataset1 = dataset1 / norms

print("\nDataset 1 (after L2 Normalization):")
print(normalized_l2_dataset1)
```

#Broadcasting and the softmax function

```python
import numpy as np

def softmax(z):
    # Ensure numerical stability by subtracting the maximum value
    z_max = np.max(z, axis=1, keepdims=True)
    exp_z = np.exp(z - z_max)  # Subtract z_max for numerical stability
    sum_exp_z = np.sum(exp_z, axis=1, keepdims=True)
    softmax_probs = exp_z / sum_exp_z
    return softmax_probs

# Example usage with broadcasting
# Generate a synthetic dataset of logits (3 samples, 4 classes)
logits = np.array([[1.0, 2.0, 3.0, 4.0],
                   [-1.0, 0.0, 1.0, 2.0],
                   [0.5, 1.5, 2.5, 3.5]])

# Compute softmax probabilities using broadcasting
softmax_probs = softmax(logits)

# Display original logits and computed softmax probabilities
print("Logits:\n", logits)
print("\nSoftmax Probabilities:\n", softmax_probs)
```

#Vectorization

```python
import numpy as np
import time

def classic_dot_product(x1, x2):
    tic = time.process_time()
    dot = 0
    for i in range(len(x1)):
        dot += x1[i] * x2[i]
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return dot, computation_time

def classic_outer_product(x1, x2):
    tic = time.process_time()
    outer = np.zeros((len(x1), len(x2)))
    for i in range(len(x1)):
        for j in range(len(x2)):
            outer[i, j] = x1[i] * x2[j]
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return outer, computation_time

def classic_elementwise_multiplication(x1, x2):
    tic = time.process_time()
    mul = np.zeros(len(x1))
    for i in range(len(x1)):
        mul[i] = x1[i] * x2[i]
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return mul, computation_time

def classic_general_dot_product(W, x1):
    tic = time.process_time()
    gdot = np.zeros(W.shape[0])
    for i in range(W.shape[0]):
        for j in range(len(x1)):
            gdot[i] += W[i, j] * x1[j]
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return gdot, computation_time

def vectorized_dot_product(x1, x2):
    tic = time.process_time()
    dot = np.dot(x1, x2)
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return dot, computation_time

def vectorized_outer_product(x1, x2):
    tic = time.process_time()
    outer = np.outer(x1, x2)
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return outer, computation_time

def vectorized_elementwise_multiplication(x1, x2):
    tic = time.process_time()
    mul = np.multiply(x1, x2)
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return mul, computation_time

def vectorized_general_dot_product(W, x1):
    tic = time.process_time()
    gdot = np.dot(W, x1)
    toc = time.process_time()
    computation_time = 1000 * (toc - tic)  # Convert time to milliseconds
    return gdot, computation_time

# Sample data
x1 = [9, 2, 5, 0, 0, 7, 5, 0, 0, 0, 9, 2, 5, 0, 0]
x2 = [9, 2, 2, 9, 0, 9, 2, 5, 0, 0, 9, 2, 5, 0, 0]
W = np.random.rand(3, len(x1))

# Perform computations
dot_classic, time_classic_dot = classic_dot_product(x1, x2)
outer_classic, time_classic_outer = classic_outer_product(x1, x2)
elementwise_classic, time_classic_elementwise = classic_elementwise_multiplication(x1, x2)
gdot_classic, time_classic_gdot = classic_general_dot_product(W, x1)

dot_vectorized, time_vectorized_dot = vectorized_dot_product(x1, x2)
outer_vectorized, time_vectorized_outer = vectorized_outer_product(x1, x2)
elementwise_vectorized, time_vectorized_elementwise = vectorized_elementwise_multiplication(x1, x2)
gdot_vectorized, time_vectorized_gdot = vectorized_general_dot_product(W, x1)

# Print results
print("Classic Dot Product:")
print("dot = {}, Computation time = {} ms".format(dot_classic, time_classic_dot))
print("\nClassic Outer Product:")
print("outer = {}, Computation time = {} ms".format(outer_classic, time_classic_outer))
print("\nClassic Elementwise Multiplication:")
print("elementwise multiplication = {}, Computation time = {} ms".format(elementwise_classic, time_classic_elementwise))
print("\nClassic General Dot Product:")
print("gdot = {}, Computation time = {} ms".format(gdot_classic, time_classic_gdot))

print("\nVectorized Dot Product:")
print("dot = {}, Computation time = {} ms".format(dot_vectorized, time_vectorized_dot))
print("\nVectorized Outer Product:")
print("outer = {}, Computation time = {} ms".format(outer_vectorized, time_vectorized_outer))
print("\nVectorized Elementwise Multiplication:")
print("elementwise multiplication = {}, Computation time = {} ms".format(elementwise_vectorized, time_vectorized_elementwise))
print("\nVectorized General Dot Product:")
print("gdot = {}, Computation time = {} ms".format(gdot_vectorized, time_vectorized_gdot))
```

#Implement the L1 and L2 loss functions

image.png

```python
def l1_loss(y_true, y_pred):
    n = len(y_true)
    loss = sum(abs(y_true[i] - y_pred[i]) for i in range(n)) / n
    return loss
```

image.png

```python
def l2_loss(y_true, y_pred):
    n = len(y_true)
    loss = sum((y_true[i] - y_pred[i]) ** 2 for i in range(n)) / n
    return loss
```

```python
# Example usage
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

# Calculate L1 loss (MAE)
l1_error = l1_loss(y_true, y_pred)
print("L1 Loss (MAE):", l1_error)

# Calculate L2 loss (MSE)
l2_error = l2_loss(y_true, y_pred)
print("L2 Loss (MSE):", l2_error)
```
